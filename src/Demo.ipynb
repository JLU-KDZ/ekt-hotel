# mock

```python
from unittest.mock import patch

def my_print():
    return "hello!"

with patch("__main__.my_print") as mock_my_print:
    mock_my_print.return_value= "world!"
    print(my_print())
    
print(my_print())
```

```python
from unittest.mock import patch

class MyC:
    def __init__(self):
        print("init!")
    def my_print(self):
        print("hello!")

# 但 patch 需要提供 完整的模块路径，即 MyC 所在的 模块名.类名.方法名。
with patch("__main__.MyC.my_print") as mock_MyC_my_print:
    mock_MyC_my_print.return_value= "world!"
    myc=MyC()
    print(myc.my_print())

with patch("__main__.MyC") as mock_MyC:
    
    print("mock_MyC type: ",type(mock_MyC))
    print("mock_MyC.return_value type: ",type(mock_MyC.return_value))
    
    # 替换了整个 MyC 类，所以当你调用 MyC() 时得到的是 mock_MyC.return_value。要设置返回值，应配置 mock_MyC.return_value.my_print.return_value。在第二个 patch 中，你实际上是在替换整个类 MyC。当你这样替换后，调用 MyC() 会返回一个 MagicMock 实例，而这个实例是 mock_MyC.return_value。
    # mock_MyC.my_print.return_value = "new world!" 这个写法是错误的。
    mock_MyC.return_value.my_print.return_value = "new world!"
    myc=MyC()
    print(myc.my_print())
```
